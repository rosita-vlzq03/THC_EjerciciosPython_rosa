```python
"""
  Programe una función que calcule el área de un triángulo utilizando la fórmula de Herón.
  Complemente el ejercicio con las condiciones solicitadas en la tarea 3:
  1.- Que los valores del triángulo sean positivos. Que se repita el programa hasta los tres valores sean positivos.
  (hint: ciclo while y condicional if-elif-else).
  2.- Que se cumpla la desigualdad triangular.
  3.- Operaciones (calculo del semiperímetro y el área).
  4.- Indique con comentarios sus variables globales y locales.
  5.- Llamada a la función con sus argumentos e impresión en pantalla del resultado.

"""

#Definimos la función con tres parámetros
def heron(a,b,c): #La función calcula el semiperímetro y el area, siempre y cuando los valores ingresados por el usuario sean válidos
  s=(a+b+c)/2 #Variable local para el semiperímetro
  area_triang=(s*(s-a)*(s-b)*(s-c))**0.5 #Variable local para el área con la fórmula de Herón
  return s, area_triang #La función devuelve dos valores



while True:
  a=int(input("Ingrese el lado a de un triángulo: ")) #a, b y c son variables globales, pues no están dentro de la función
  b=int(input("Ingrese el lado b de un triángulo: "))
  c=int(input("Ingrese el lado c de un triángulo: "))

  if a<=0 or b<=0 or c<=0: #Condición de que al menos un lado es negativo
    print("Longitudes no válidas, las tres longitudes deben ser positivas")
    continue #Si se cumple esto, entonces continue hace que el programa regrese al while
  elif a+b>c and a+c>b and b+c>a: #Si los lados son positivos, se verifica la desigualdad triangular
    print("Los valores ingresados sí forman un triángulo")
    s,area=heron(a,b,c) #Si se cumple la desigualdad triangular, se llama a la función,
                        #guardando cada valor que devuelve la función en las variables globales s y area,
                        #en el orden en el que la función devuelve los valores"""
    print(f"El semiperímetro del triángulo es: {s}") #Se imprime el semiperímetro
    print(f"El área del triángulo es: {area}") #Se imprime el área
    break #Como todos los valores fueron positivos y si formaron un triángulo, se rompe el ciclo while
  else: #Si no se cumplen el if y elif anteriores, entonces se imprime en pantalla que a, b y c no forman un triángulo
    print("Los valores ingresados no forman un triángulo")
    break #Si a, b y c no forman un triángulo, entonces se rompe el ciclo while
```
