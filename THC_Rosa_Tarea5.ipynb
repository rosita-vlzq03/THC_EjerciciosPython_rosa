Ejercicio 1.- Ingreso a una boveda.

Pedro y Jesús son dos hermanos, hijos de Manuel, y quienes tienen acceso a una bóveda en un exclusivo banco. El dueño original de esta bóveda es su padre, y en su testamento ha especificado las siguientes condiciones para que el banco le permita acceder a sus hijos al contenido de dicha bóveda. Las condiciones son las siguientes:

a. Juan debe ser mayor de 35 años, o Pedro debe ser menor de 50 años.

b. La suma de sus edades debe ser un número par.

c. Manuel debe tener mínimo 1 nieto.

Construya una función que lea la edad de Juan, la edad de Pedro, el número de nietos y calcule, usando condicionales, si tienen acceso (SI) o no (NO) a la bóveda.

```python
def abrir_boveda(juan,pedro,num_nietos): #Definimos la función con tres parámetros, las edades de Juan y Pedro, y el número de nietos de Manuel.
  if juan>35 or pedro<50: #Si se cumple la primera condición, se suman las edades, si no, la función regresa 'NO'
    suma_edad=juan+pedro
    if suma_edad%2==0: #Se analiza la suma anterior, si es par, el programa continúa, si no, la función regresa 'No'
      if num_nietos>=1: #Si hay al menos un nieto, la función regresa 'SÍ', porque ya se han cumplido todas las condiciones, si no, la función regresa 'NO'
        return "SI"
      else: return "NO"
    else: return "NO"
  else: return "NO"
abrir_boveda(36,44,5) #Se llama a la función con algunos valores
```

Ejercicio 2.- Reloj mundial.

Utilizando listas, diccionarios y fuciones Lambda, programe un reloj mundial, en el que ingrese como entrada la hora de la Ciudad de México y muestre la hora en las siguientes ciudades.

1.- Tokio, Japón

2.- Pekin, China

3.- Dusambé, Tayikistán

4.- Perth, Australia

5.- Katmandú, Nepal

6.- Vilna, Lituania

7.- Freetown, Sierra Leona

8.- Río de Janeiro, Brasil

9.- Tegucigalpa, Honduras

10.- La Paz, Baja California Sur, México

Comentar los procesos esenciales del código.

```python
#El UTC de CDMX respecto del Meridiano Cero
cdmx_utc=-6

#Diccionario con las ciudades dadas y el UTC, correspondiente
cities_utc={"Tokio, Japón":+9, "Pekin, China":+8, "Dusambé, Tayikistán":+5,"Perth, Australia":+8,"Katmandú,Nepal":+5.75,"Vilna,Lituania":+2,
            "Freetown, Sierra Leona":+0,"Río de Janeiro, Brasil":-3, "Tegucigalpa,Honduras":-6,"La Paz, BCS,México":-7}

#Defino una función para ingresar la hora actual en la Ciudad de México llamando a la función
def hora_cdmx(hour, min):
  if 0<=hour<=23 and 0<=min<=59:
    return hour,min
  else:
    raise ValueError("Hora o minutos fuera de rango")

h_cdmx, m_cdmx = hora_cdmx(7, 31) #Llamamos a la función de la hora de cdmx con la hora actual

#Imprimimos primero la hora en CDMX
print(f"Hora actual en Ciudad de México: {h_cdmx:02d}:{m_cdmx:02d}\n")
print(f"Cuando en Ciudad de México son las {h_cdmx:02d}:{m_cdmx:02d}, en las siguientes ciudades es:\n")

#Para cada ciudad en el diccionario, calculamos la diferencia de horario respecto de la Ciudad de México
for ciudad,utc in cities_utc.items():
  diferencia=utc-cdmx_utc

  #En h*60+m convertimos las horas a minutos y le sumamos la diferencia de utc de la ciudad con  el utc de cdmx igual en minutos,
  #a ese resultado hacemos la división entera con 60, para que nos dé las horas completas, y tomamos el módulo al dividir entre 24
  #así sabemos si la diferencia de horas es menor o mayor a un día
  #para los minutos hacemos el mismo cálculo pero en vez de hacer división entera, obtenemos el módulo al dividir entre 60 y así obtenemos
  #el valor final para los minutos

  calcular_hora = lambda h, m: (
        int(((h * 60 + m) + diferencia * 60) // 60) % 24,
        int(((h * 60 + m) + diferencia * 60) % 60)
    )

  h_local, m_local = calcular_hora(h_cdmx, m_cdmx) #LLamamos a la función lambda guardando en dos variables los valores obtenidos para la hora y minuto de cada ciudad

  print(f"{ciudad}: {h_local:02d}:{m_local:02d}") #Cada vez que el ciclo for lee cada clave y valor en el diccionario y se hacen todos los cálculos
                                                  #al final imprimimos la hora de la ciudad que se está leyendo
```

Ejercicio 3.- Número primo.

Con base en la definición de número primo, escriba una función que reciba un número entero e imprima 'True' si lo es o de lo contrario imprima 'False'. Tenga en cuenta que si el número es menor o igual a 1, se retorna como resultado "Error en el Número" y debe volver a preguntar hasta que la entrada sea valida.

```python
def primo(n): #Definimos una función que analiza si un entero positivo mayor a 1 es primo o no
  if n<=1: #Si n es menor o igual a 1 se devuelve un error
    raise ValueError("Error en el número")
  elif n%n==0 and n%1==0: #Si n es mayor que 1, se pide que su división entre sí mismo y entre 1, tenga módulo cero
    for i in range (2,n): #Como todo número entero n cumple la condición anterior, utilizamos el ciclo for para que analize si es divisible
      if n%i==0:          #también entre los números comprendidos entre n y 2, si encuentra algún divisor, entonces la función regresa False
        return False
    else: return True     #Si el ciclo for no encuentra más divisores, entonces la función regresa True

primo(2)
```

Ejercicio 5.- MCD y MCM.

Dados dos números a y b, determine:

El máximo común divisor.
El mínimo común múltiplo.
Almacene los resultados en un diccionario e imprima en pantalla el contenido del diccionario.

```python
import math #Importamos el módulo math

min_maxab={} #Creamos un diccionario vacío

def mult_div(a,b): #Definimos la función para calcular mcm y MCD
  if a>0 and b>0: #Si ambos números son positivos, con la función math.gcd calculamos el máximo común divisor
    MCD=math.gcd(a,b)
    min_maxab["MCD"]=MCD #Añadimos el máximo común divisor al diccionario
    mcm=a*b//MCD #Con el MCD calculamos el mínimo común múltiplo
    min_maxab["mcm"]=mcm #Añadimos el mínimo común múltiplo al diccionario
  else: #Si los números son negativos o cero devolvemos un error
    raise ValueError("Los números deben ser positivos")

mult_div(6,8) #llamamos a la función con dos valores

print(f"El máximo común divisor y el mínimo común múltiplo de los números dados son {min_maxab} respectivamente.") #Imprimimos el nuevo diccionario

```

Ejercicio 6.- Filtrado de datos.

Escriba una función que reciba una lista de datos y devuelva listas con los datos filtrados. Este programa debe separar los datos en int, str, bool, float.

```python
datos = [25, "Matemáticas", 3.141592, True, 8, False, "Nombres", 2.718, "UNAM", 0,
         9.81, "CdMx", 7e12, 12.5, "2025", True, "población", 0.01, 42, False]

def datos_filtro(datos): #Definimos la función que recibe la lista datos como parámetro

  enteros=[] #Creamos una lista vacía por cada tipo de dato
  cadenas=[]
  booleano=[]
  flotante=[]

  for elemento in datos: #Ciclo for para cada elemento en la lista dada, guardamos en la variable tipo, el tipo de elemento de la lista datos
    tipo=type(elemento)
    if tipo==str: #Con el condicional if-elif-elif-else cubrimos cada caso según el tipo de elemento de la lista datos y según el tipo, se agrega
      cadenas.append(elemento)   #con lista.append en cada una de las listas vacías
    elif tipo==int:
      enteros.append(elemento)
    elif tipo==bool:
      booleano.append(elemento)
    else:
      flotante.append(elemento)
  print(cadenas)
  print(enteros)
  print(booleano)
  print(flotante)

datos_filtro(datos)
```

Realice un programa que calcule la serie de Fibonacci y la imprima en pantalla, tenga en cuenta las restricciones de la variable de entrada.

```python
fib_0=1 #Condiciones iniciales de la sucesión recurrente
fib_1=1

n=int(input("Ingrese un número entero mayor o igual a 2: ")) #pedimos al usuario un número, el cual determinará la cantidad de términos a imprimir

while n<2: #Si el número ingresado es menor a dos, se le pedirá al usuario un valor hasta que ingrese uno mayor o igual a 2
  print("Número no válido")
  n=int(input("Ingrese un número entero mayor o igual a 2: "))

print(f"Los primeros {n} términos de la sucesión de Fibonacci son: ") #Imprimimos los dos primeros términos
print(f"f0={fib_0}")
print(f"f1={fib_1}")

for i in range(1,n-1): #Con el ciclo for encontramos los demás términos de la sucesión
  fib_n=fib_0+fib_1 #fib_n está determinado por las condiciones iniciales
  fib_0=fib_1 #Recorremos un término hacia adelante cada término, así cambiara en cada iteración de i
  fib_1=fib_n

  print(f"f{i+1}={fib_n}") #el ciclo for imprime cada término en cada iteración
```

Ejercicio 9.- Mini-catálogo de sismos.

En este ejercicio se muestran 4 listas que representan un conjunto de sismos con la siguiente información (fecha, zona, magnitud y profundidad). Hacer un programa que haga lo siguiente:

1.- Convierte la lista de magnitudes de str a float.

2.- Los nombres de los estados deben iniciar con mayúscula (utilice una función lambda).

3.- Crea una lista llamada intensidad (con la misma longitud de la lista de magnitudes) y coloca etiquetas de acuerdo al siguiente criterio:

* Magnitud < 3.5 "Muy Leve"
* 3.5 <= Magnitud < 4.5 "Leve"
* 4.5 <= Magnitud < 3.5 "Moderado"
* Magnitud > 5.5 "Fuerte"

4.- Con un ciclo For (y condicionales si es necesario), genera un diccionario "conteo_por_zona" con el número de eventos por zona.

5.- Ordena los eventos por magnitud descendente y muestra los 3 sismos más grandes con el siguiente formato:

YYYY-MM-DD | Zona (Estado) | Magnitud | Profundidad (km).

```python
Fechas = ["2025-03-01","2025-03-02","2025-03-03","2025-03-04","2025-03-05",
    "2025-03-06","2025-03-07","2025-03-08","2025-03-09","2025-03-10"]

Zonas = ["guerrero","oaxaca","chiapas","michoacán","guerrero",
    "oaxaca","chiapas","guerrero","oaxaca","chiapas"]

magnitudes = ["3.8","4.2","5.0","2.9","4.8","3.5","4.1","5.3","4.0","3.0"]

prof_km = [15, 80, 12, 5, 35, 120, 60, 25, 10, 90] #Profundidad en kilómetros.

#Con un ciclo for tomamos todos los elementos de magnitudes y los convertimos en floats
i=0
for m in magnitudes:
  decimal_magn=float(m)
  magnitudes[i]=decimal_magn
  i+=1

#Con una función lambda, creamos una nueva lista cambiando la inicial de cada estado en la lista Zonas por una mayúscula
Zonas_mayus=list(map(lambda estado:estado[0].upper()+ estado[1:],Zonas))

#Lista vacia de la intensidad de los sismos
intensidad=[]

#Con este ciclo for, para cada elemento en magnitudes, dependiendo en qué rango está, agregamos un string en la lista intensidad, de acuerdo a las etiquetas dadas
for m in magnitudes:
  if m<3.5:
    intensidad.append("Muy Leve")
  elif 3.5<=m<4.5:
    intensidad.append("Leve")
  elif 4.5<=m<5.5:
    intensidad.append("Moderado")
  elif m>5.5:
    intensidad.append("Fuerte")

#Diccionario vacío para la recurrencia de sismos por estado
conteo_por_zona={}

#Con el ciclo for, iteramos sobre cada estado en la lista Zona_mayus, y si estado está en el diccionario, se le asigna un valor, dependiendo de cuantas veces aparece
#en la lista Zona_mayus
for estado in Zonas_mayus:
  if estado in conteo_por_zona:
    conteo_por_zona[estado]+=1
  else:
    conteo_por_zona[estado]=1

#Con la función zip, creamos tuplas con cada parámetro de los sismos, y estas tuplas son los elementos de la lista eventos
eventos=list(zip(Fechas,Zonas_mayus,magnitudes,prof_km))

#Creamos una nueva lista a partir de la lista eventos, donde con la función sorted para los elementos de eventos, los ordenamos en orden mayor a menor de acuerdo al
#parámetro con índice [2], es decir la magnitud del sismo
eventos_ord=sorted(eventos,key=lambda eventos:eventos[2], reverse=True)

#Como ya están ordenados de mayor a menor, solo basta con tomar los tres primeros elementos de la lista eventos_ord
top3_sismos=eventos_ord[:3]

#Imprimimos todas las listas y diccionarios solicitados en las instrucciones
print(f"\nLa lista magnitudes con sus elementos convertidos en tipo float es: {magnitudes}")
print(f"\nLa lista Zonas con las iniciales de los estados en mayúsculas es: {Zonas_mayus}")
print(f"\nLa lista intensidad etiqueta cada sismo de acuerdo a su magnitud y es {intensidad}")
print(f"\nEl diccionario conteo_por_zona muestra la recurrencia de sismos en diferentes estados y es {conteo_por_zona}")
print(f"\nLa lista eventos_ord muestra los datos de cada sismo y los ordena en forma descendente y es {eventos_ord}")

#Con un ciclo for imprimimos cada parámetro de las tuplas en la lista top3_sismos
print("\nLos tres sismos más grandes son: ")
for fecha,zona,magnitud,profundidad in top3_sismos:
  print(f"{fecha} | {zona} | {magnitud} | {profundidad}km")
```

Ejercicio 10.- Operaciones vectoriales.

Dadas las siguientes listas:

a = [2, -1, 4],  b = [4, -2, 8],  c = [-5, 2, 1]

Calcule el producto escalar:
* ab
* bc
* ca

Calcular el productor vectorial:
* a x b
* b x c
* a x c

Utilice las listas de inicio para probar su código. Posteriormente, genere una función que reciba como entrada las listas (vectores) y haga las operaciones vectoriales.

Indique si algunos de los vectores son paralelos o perpendiculares.

```python
import numpy as np

#Listas iniciales
a = [2, -1, 4]
b = [4, -2, 8]
c = [-5, 2, 1]

#Función que recibe como parámetros dos listas cualesquiera
def op_vectores(v1,v2):

  #Al ingresar dos listas en la función, la función las convierte en vectores/arrays
  v1=np.array(v1)
  v2=np.array(v2)

  #producto escalar de v1 y v2
  escalar=np.dot(v1,v2)

  #producto vectorial de v1 y v2
  vectorial=np.cross(v1,v2)

  print(f"El producto escalar de {v1} y {v2} es {escalar}")
  print(f"El producto vectorial de {v1} y {v2} es {vectorial}")

  #Ortogonalidad de v1 y v2
  if np.isclose(escalar,0):
    print(f"Los vectores {v1} y {v2} son ortogonales")
  #Vectores paralelos si el producto cruz es el vector nulo
  elif np.allclose(vectorial,[0,0,0]):
    print(f"Los vectores {v1} y {v2} son paralelos")
  else:
    print(f"Los vectores {v1} y {v2} no son ortogonales ni paralelos")
  print("-"*50)

op_vectores(a,b)

op_vectores(b,c)

op_vectores(a,c)


```
